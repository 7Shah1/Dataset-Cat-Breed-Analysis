```python
from icrawler.builtin import BingImageCrawler
import os

# 10 additional cat breeds
cat_breeds = [
    'turkish van',
    'oriental shorthair',
    'savannah cat',
    'russian blue',
    'devon rex',
    'birman cat',
    'american shorthair',
    'manx cat',
    'chartreux',
    'balinese cat'
]

# Create image folders and crawl up to 300 images per breed
for breed in cat_breeds:
    folder_name = breed.replace(" ", "_").lower()
    save_dir = os.path.join('images', folder_name)
    os.makedirs(save_dir, exist_ok=True)

    print(f"Downloading images for: {breed} â†’ {save_dir}")
    crawler = BingImageCrawler(storage={'root_dir': save_dir})
    crawler.crawl(keyword=breed, max_num=300)

print("\nâœ… Download complete.")

```

```python
import os
from PIL import Image
import imagehash

def remove_duplicates(directory):
    print(f"Checking for duplicates in {directory}...")
    hashes = set()
    removed = 0

    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        try:
            with Image.open(file_path) as img:
                img_hash = imagehash.average_hash(img)
                if img_hash in hashes:
                    os.remove(file_path)
                    removed += 1
                else:
                    hashes.add(img_hash)
        except Exception as e:
            print(f"Error with {file_path}: {e}")
            continue

    print(f"Removed {removed} duplicate images from {directory}")

# Run for each class directory
root_dir = 'images'
for class_folder in os.listdir(root_dir):
    class_dir = os.path.join(root_dir, class_folder)
    if os.path.isdir(class_dir):
        remove_duplicates(class_dir)
```
